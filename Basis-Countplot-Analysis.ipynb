# Exploring Mushrooms

Mushrooms exist in a variety of different colors, shapes, sizes, textures, etc. In this project, we will analyze an extensive mushroom dataset from UCI using bar charts and acquaint ourself with the diverse array of mushrooms that exist worldwide.



```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('mushroom_data.csv') 
print(df.columns)
df.head()
```

```python
for column in df.columns:
    plt.figure(figsize=(10,8))
    sns.countplot(df[column],order=df[column].value_counts().index)
    plt.title(f'Countplot For {column}')
    plt.show()
    plt.clf()
```
