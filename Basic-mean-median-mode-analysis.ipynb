# Central Tendency for Housing Data
In this project, we will find the mean, median, and mode cost of one-bedroom apartments in three of the five New York City boroughs: Brooklyn, Manhattan, and Queens.

Using our findings, we will make conclusions about the cost of living in each of the boroughs. We will also discuss an important assumption that we make when we point out differences between the boroughs.

```python
import pandas as pd
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

```python
# Read in housing data
brooklyn_one_bed = pd.read_csv('brooklyn-one-bed.csv')
brooklyn_price = brooklyn_one_bed['rent']
manhattan_one_bed = pd.read_csv('manhattan-one-bed.csv')
manhattan_price = manhattan_one_bed['rent']

queens_one_bed = pd.read_csv('queens-one-bed.csv')
queens_price = queens_one_bed['rent']
```

```python
# Finnding the average value of all the apartments in all 3 boroughs
brooklyn_mean = np.average(np.array(brooklyn_price))
print(f'Average price of one Bedrrom apartment in Brooklyn is ${brooklyn_mean}')

manhattan_mean = np.average(np.array(manhattan_price))
print(f'Average price of one Bedrrom apartment in Manhattan is ${manhattan_mean}')

queens_mean = np.average(np.array(queens_price))
print(f'Average price of one Bedrrom apartment in Queens is ${queens_mean}')
```

```python
# Finnding the median value of all the apartments in all 3 boroughs
brooklyn_median = np.median(np.array(brooklyn_price))
print(f'Median price of one Bedrrom apartment in Brooklyn is ${brooklyn_median}')

manhattan_median= np.median(np.array(manhattan_price))
print(f'Median price of one Bedrrom apartment in Manhattan is ${manhattan_median}')

queens_median= np.median(np.array(queens_price))
print(f'Median price of one Bedrrom apartment in Queens is ${queens_median}')
```

```python
# Finnding the mode value of all the apartments in all 3 boroughs
brooklyn_mode = stats.mode(np.array(brooklyn_price))
print(f'Mode price of one Bedrrom apartment in Brooklyn is ${brooklyn_mode[0]}')

manhattan_mode= stats.mode(np.array(manhattan_price))
print(f'Mode price of one Bedrrom apartment in Manhattan is ${manhattan_mode[0]}')

queens_mode= stats.mode(np.array(queens_price))
print(f'Mode price of one Bedrrom apartment in Queens is ${queens_mode[0]}')
```

It looks like the average cost of one-bedroom apartments in Manhattan is the most, and in Queens is the least. This pattern holds for the median and mode values as well.

While the mode is not the most important indicator of centrality, the fact that mean, median, and mode are within a few hundred dollars for each borough indicates the data is centered around:

$3,300 for Brooklyn,

$3,900 for Manhattan,

$2,300  for Queens.
